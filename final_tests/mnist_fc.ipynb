```python
import matplotlib
matplotlib.use('Agg')

import numpy as np
import cPickle
import os, sys
import tensorflow as tf
import matplotlib.pyplot as plt
from rbm import RBM
from crbm import CRBM
import train_rbm
import dbm as dbm_class
```

```python
# Load data
(train_xs, _), _, _ = cPickle.load(file('mnist.pkl', 'rb'))
batch_size = 100
# Set parameters
lr = 0.001
output_dir = "fc_2L_batch_100"
#output_dir = "fc_2L"
#output_dir = "fc_deep"
#output_dir = "fc_3L_small_batch"
```

```python
dbm = dbm_class.DBM([784])
dbm.add_fc_layer(500, 'fc1')
dbm.add_fc_layer(1000, 'fc2')
#dbm.add_fc_layer(1000, 'fc3')
dbm.print_network()
```

```python
train_rbm.train(dbm, train_xs, lr, 2000, batch_size, True, 5, output_dir,
                steps_per_output=100, use_adam=False,
                lr_decay=True, decay_steps=50,
                save_model=True, steps_per_save=100)
```

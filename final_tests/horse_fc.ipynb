```python
import matplotlib
matplotlib.use('Agg')

import numpy as np
import cPickle
import os, sys
import tensorflow as tf
import matplotlib.pyplot as plt
from rbm import RBM
from crbm import CRBM
import train_rbm
import dbm as dbm_class
```

```python
# Load data
train_set = cPickle.load(file('horse_small.pkl', 'rb'))
# flatten the data
train_set = train_set.reshape(327,784)
# ignore last few images
train_set = train_set[0:300,:]
batch_size = 10
# Set parameters
lr = 0.002
output_dir = "horse_2L_batch_10_small"
```

```python
dbm = dbm_class.DBM([784])
dbm.add_fc_layer(500, 'fc1')
dbm.add_fc_layer(1000, 'fc2')
dbm.print_network()
```

```python
train_rbm.train(dbm, train_set, lr, 4000, batch_size, True, 5, output_dir,
                steps_per_output=10, use_adam=False,
                lr_decay=True, decay_steps=150,
                save_model=True, steps_per_save=100)
```
